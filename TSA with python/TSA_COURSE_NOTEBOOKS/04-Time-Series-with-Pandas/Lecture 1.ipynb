```python
airline = pd.read_csv('../Data/airline_passengers.csv',index_col='Month',parse_dates=True)
```

```python
airline.dropna(inplace=True)
```

```python
airline.head()
```

```python
airline.rename(columns = {'Thousands of Passengers':'Passengers'},inplace=True)
```

```python
airline.head()
```

```python
airline.plot();
```

```python
airline['12-month-SMA'] = airline['Passengers'].rolling(window=12).mean()
airline['12-month-SMA'].plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(airline['Passengers'], model='multiplicative')  # model='mul' also works
result.plot();
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(airline['Passengers'], model='additive')  # model='mul' also works
result.plot();
```

```python
airline['d1']=airline.Passengers.diff()
```

```python
result = seasonal_decompose(airline['d1'], model='mul')  # model='mul' also works
result.plot();
```

```python
airline.dropna(inplace=True)
```

```python
result = seasonal_decompose(airline['d1'], model='mul')  # model='mul' also works
result.plot();
```

```python
result = seasonal_decompose(airline['d1'], model='add')  # model='mul' also works
result.plot();
```
